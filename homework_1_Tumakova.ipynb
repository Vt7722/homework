### 1. Как вы крастко охарактеризуете датасет, про что он?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

dataset_filename = "Google-Playstore.csv"
data_frame = pd.read_csv(dataset_filename)
data_frame
```

Данный датасет про приложения в "Google Play". Он содержит в себе более двух миллионов строк и 23 атрибута. Из основных атрибутов можно выделить следующие: 
* Название приложения

* Количество установок
* Цена
* Размер

## 2. Сколько в среднем скачиваний у приложений в выборке?

```python
data_frame["Installs"]=data_frame["Installs"].str.replace(",","")
data_frame["Installs"]=data_frame["Installs"].str.replace("+","", regex=True)
df = data_frame["Installs"].dropna().astype(dtype="Int64")
from scipy.stats import trim_mean
print("Среднее арифметическое - ", df.mean() , "\n", 
      "Среднеусеченное - ", trim_mean(df, 0.15), "\n",
      "Медиана - ", df.median(axis=0, skipna=True, numeric_only=True), "\n", 
      "Мода - ", df.mode(), "\n" )
```

Среднее арифметическое равно 183445, среднеусеченное - 1833

## 3. Сколько в среднем стоит приложение в выборке?

```python
df = data_frame["Price"].dropna()
from scipy.stats import trim_mean
print("Среднее арифметическое - ", df.mean() , "\n", 
      "Среднеусеченное - ", trim_mean(df, 0.15), "\n",
      "Медиана - ", df.median(axis=0, skipna=True, numeric_only=True), "\n", 
      "Мода - ", df.mode(), "\n" )
df = df[df > 0.0]
print("Среднее арифметическое - ", df.mean() , "\n", 
      "Среднеусеченное - ", trim_mean(df, 0.15), "\n",
      "Медиана - ", df.median(axis=0, skipna=True, numeric_only=True), "\n", 
      "Мода - ", df.mode(), "\n" )
```

Если брать все значения включая цену 0, то среднее арифметическое равно 0.1, а среднеусеченное равно 0. Часто встречаемое, т.е мода, это 0. Если брать значения отличные от 0, то среднее арифметическое равно 5.32, среднеусеченное равно 2.63, а часто встречаемое значение - 0.99.

## 4. Как распределены данные по скачиваниям?

```python
data_frame["Installs"]=data_frame["Installs"].str.replace(",","")
data_frame["Installs"]=data_frame["Installs"].str.replace("+","", regex=True)
df = data_frame["Installs"].dropna().astype(dtype="Int64")
plt.subplot (2, 1 ,1)
plt.boxplot(df, vert=False)
plt.subplot(2, 1, 2)
plt.boxplot(df[df<5000], vert=False)
plt.show()
```

Данные по скачиваниям имеют положительную ассиметрию. На верхнем рисунке изображены все данные. На нижнем часть данных без выбросов.

## 10. Как кореллируют между собой рейтинг и количество скачиваний приложения?

```python
import seaborn as sns 
data_frame["Installs"] = data_frame["Installs"].astype(float)
c = data_frame[["Rating", "Installs"]].corr()
sns.heatmap(c, cmap="BrBG", annot=True)
c
```
